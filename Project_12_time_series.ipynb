На будущее: можно проверить ряд на стационарность с помощью теста Дики-Фуллера `adfuller()`
    
Стационарный ряд - это такой ряд, у которого среднее и стандартное отклонение не меняются.

Проверка на стационарность может быть выполнена следующими способами:
- Визуально: отобразать временной ряд и проверить тенденции или сезонность
- Статистически: разделить временные ряды и сравнить среднее значение и дисперсию каждого раздела
- Провести статистический тест: тест Дики-Фуллера
    
Расширенный Тест Дики — Фуллера, 4 версии теста:
- Без константы и тренда: 'nc' 
- С константой, но без тренда: 'c' 
- С константой и линейным трендом: 'ct' 
- С константой, линейным и квадратичным трендом: 'ctt'
    
    
Пример: 
    
        from statsmodels.tsa.stattools import adfuller
    
        ts = taxi['num_orders']
        st_test = adfuller(ts, regression='ctt')

## Обучение

Создадим дополнительные признаки часа, месяца, дня месяца и дня недели

```python
def make_features(data, max_lag, rolling_mean_size):
    data['hour'] = data.index.hour
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()


make_features(df_resH, 24, 12)
```

Разделим на обучающую и тестовую выборки

```python
train_base, test = train_test_split(df_resH, shuffle=False, test_size=0.1)
train, valid = train_test_split(train_base, shuffle=False, test_size=0.1)
train = train.dropna()
valid = valid.dropna()
train.info()
valid.info()
test.info()
```

Выделим целевые признаки

```python
target_train = train['num_orders']
features_train = train.drop('num_orders', axis=1) 

target_valid = valid['num_orders']
features_valid = valid.drop('num_orders', axis=1) 

target_test = test['num_orders']
features_test = test.drop('num_orders', axis=1)
```

Рассмотрим модели DecisionTreeRegressor, LinearRegression, Ridge, CatBoostRegressor

Здорово, что при делении на выборки указан параметр `shuffle=False`, т.к. в `train_test_split` его дефолтное значение True. В случае временных рядов выборки перемешивать не стоит 👍

### DecisionTreeRegressor

```python
%%time

model_DTR = DecisionTreeRegressor() 
max_depth_list = [x for x in range(2, 21)]
param_grid = [{'criterion':['mse'], 
                'max_depth':max_depth_list, 
                'random_state':[RANDOM_STATE]}]
tscv = TimeSeriesSplit(n_splits=2)
clf = GridSearchCV(model_DTR, param_grid, cv=tscv, scoring='neg_mean_squared_error')
clf.fit(features_train, target_train)
print("Best parameters:")
best_params_DTR = clf.best_params_
print(best_params_DTR)
```

```python
%%time

model_DTR = DecisionTreeRegressor()
model_DTR.set_params(**best_params_DTR)
model_DTR.fit(features_train, target_train)
RMSE_DTR = mean_squared_error(target_valid, model_DTR.predict(features_valid)) ** 0.5
print('RMSE DecisionTreeRegressor =', RMSE_DTR)
```

### LinearRegression

```python
%%time

model_LR = LinearRegression()
param_grid = {'fit_intercept': [True, False], 'normalize': [True, False]}
tscv = TimeSeriesSplit(n_splits=2)
clf = GridSearchCV(model_LR, param_grid, cv=tscv, scoring='neg_mean_squared_error')
clf.fit(features_train, target_train)
print("Best parameters:")
best_params_LR = clf.best_params_
print(best_params_LR)
```

```python
%%time

model_LR = LinearRegression()
model_LR.set_params(**best_params_LR)
model_LR.fit(features_train, target_train)
RMSE_LR = mean_squared_error(target_valid, model_LR.predict(features_valid))**0.5
print('RMSE LinearRegression =',RMSE_LR)
```

### Ridge

```python
%%time

model_R = Ridge()
param_grid = [{'alpha':[0.1, 1, 10, 100, 1000],
                'solver':['auto', 'svd', 'cholesky', 'lsqr','sparse_cg']}]
tscv = TimeSeriesSplit(n_splits=2)
clf = GridSearchCV(model_R, param_grid, cv=tscv, scoring='neg_mean_squared_error')
clf.fit(features_train, target_train)
print("Best parameters:")
best_params_R = clf.best_params_
print(best_params_R)
```

```python
%%time

model_R = Ridge()
model_R.set_params(**best_params_R)
model_R.fit(features_train, target_train)
RMSE_R = mean_squared_error(target_valid, model_R.predict(features_valid))**0.5
print('RMSE Ridge =',RMSE_R)
```

### CatBoostRegressor

```python
%%time

model_CBR = CatBoostRegressor() 
param_grid = {
    'iterations': [100, 200, 300],
    'depth': [6, 8, 10],
    'learning_rate': [0.01, 0.1, 0.2],
    'random_state': [RANDOM_STATE]}
tscv = TimeSeriesSplit(n_splits=2)
clf = GridSearchCV(model_CBR, param_grid, scoring='neg_mean_squared_error',cv=tscv)
clf.fit(features_train, target_train)
print("Best parameters:")
best_params_CBR = clf.best_params_
print(best_params_CBR)
```

```python
%%time

model_CBR = CatBoostRegressor() 
model_CBR.set_params(**best_params_CBR)
model_CBR.fit(features_train, target_train)
RMSE_CBR = mean_squared_error(target_valid, model_CBR.predict(features_valid))**0.5
print('RMSE CatBoostRegressor =',RMSE_CBR)
```

Наименьшее значение RMSE =30.9 показала модель CatBoostRegressor, остановим выбор на ней.
