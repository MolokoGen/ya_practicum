### –ö–æ–Ω–≤–µ–π–µ—Ä—ã –º–æ–¥–µ–ª–µ–π

–°–æ–∑–¥–∞–¥–∏–º –∫–æ–Ω–≤–µ–π–µ—Ä –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.  
–î–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø—Ä–∏–º–µ–Ω–∏–º –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ OneHotEncoder  
–î–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø—Ä–∏–º–µ–Ω–∏–º –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ StandardScaler

```python
numeric_features = X.select_dtypes(include=['float64','int64']).columns
categorical_features = X.select_dtypes(exclude=['float64','int64']).columns

numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False))])

preprocessor = make_column_transformer(
        (numeric_transformer, numeric_features),
        (categorical_transformer, categorical_features),
        remainder='passthrough',
        verbose_feature_names_out=False
)
```

<div class="alert alert-danger">
    <s><b>üòî –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –∏—Å–ø—Ä–∞–≤–∏—Ç—å:</b> –ü—Ä–∏ –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–∏ –Ω—É–∂–Ω–æ —É–¥–∞–ª—è—Ç—å –ø–µ—Ä–≤—ã–π —Ñ–∏–∫—Ç–∏–≤–Ω—ã–π —Å—Ç–æ–ª–±–µ—Ü</s>
</div>

<div class="alert alert-info"> <b>üéì –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Å—Ç—É–¥–µ–Ω—Ç–∞:</b> –î–æ–±–∞–≤–∏–ª.</div>

<div class="alert alert-success">
<b>üëç –£—Å–ø–µ—Ö:</b> –í—Å–µ –≤–µ—Ä–Ω–æ!
</div>

–°–æ–∑–¥–∞–¥–∏–º –∫–æ–Ω–≤–µ–π–µ—Ä –¥–ª—è –∫–∞–∂–¥–æ–π —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º–æ–π –º–æ–¥–µ–ª–∏

```python
pipe_lr = Pipeline([('preprocessor', preprocessor),
                    ('LR', LogisticRegression(random_state=RANDOM_STATE))])
pipe_dt = Pipeline([('preprocessor', preprocessor),
                    ('DT',DecisionTreeClassifier(random_state=RANDOM_STATE))])
pipe_rf = Pipeline([('preprocessor', preprocessor),
                    ('RF',RandomForestClassifier(random_state=RANDOM_STATE))])
pipe_knn = Pipeline([('preprocessor', preprocessor),
                    ('KNN', KNeighborsClassifier())])
pipe_svm = Pipeline([('preprocessor', preprocessor),
                    ('SVM', svm.SVC(random_state=RANDOM_STATE))])
pipe_cbc = Pipeline([('preprocessor', preprocessor),
                    ('CBC',CatBoostClassifier(random_state=RANDOM_STATE))])
```

–ü–æ–¥–≥–æ—Ç–æ–≤–∏–º —Å–ø–∏—Å–∫–∏ –≥–∏–ø–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏

```python
param_range = [1, 2, 3, 4, 5, 6]
param_range_fl = [1.0, 0.5, 0.1]
n_estimators = [50,100,150]
learning_rates = [.1,.2,.3]
lr_param_grid = [{'LR__penalty': ['l1', 'l2'],
                   'LR__C': param_range_fl,
                   'LR__solver': ['liblinear']}]
dt_param_grid = [{'DT__criterion': ['gini', 'entropy'],
                   'DT__min_samples_leaf': param_range,
                   'DT__max_depth': param_range,
                   'DT__min_samples_split': param_range[1:]}]
rf_param_grid = [{'RF__min_samples_leaf': param_range,
                   'RF__max_depth': param_range,
                   'RF__min_samples_split': param_range[1:]}]
knn_param_grid = [{'KNN__n_neighbors': param_range,
                   'KNN__weights': ['uniform', 'distance'],
                   'KNN__metric': ['euclidean', 'manhattan']}]
svm_param_grid = [{'SVM__kernel': ['linear', 'rbf'], 
                    'SVM__C': param_range}]
cbc_param_grid = [{'CBC__learning_rate': learning_rates,
                    'CBC__depth': param_range,
                    'CBC__iterations': n_estimators}]
```

–î–ª—è –ø–æ–∏—Å–∫–∞ –ª—É—á—à–∏—Ö –≥–∏–ø–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –ø—Ä–∏–º–µ–Ω–∏–º GridSearchCV

```python
lr_grid_search = GridSearchCV(estimator=pipe_lr,
        param_grid=lr_param_grid,
        scoring='roc_auc',
        cv=3)
dt_grid_search = GridSearchCV(estimator=pipe_dt,
        param_grid=dt_param_grid,
        scoring='roc_auc',
        cv=3)
rf_grid_search = GridSearchCV(estimator=pipe_rf,
        param_grid=rf_param_grid,
        scoring='roc_auc',
        cv=3)
knn_grid_search = GridSearchCV(estimator=pipe_knn,
        param_grid=knn_param_grid,
        scoring='roc_auc',
        cv=3)
svm_grid_search = GridSearchCV(estimator=pipe_svm,
        param_grid=svm_param_grid,
        scoring='roc_auc',
        cv=3)
cbc_grid_search = GridSearchCV(estimator=pipe_cbc,
        param_grid=cbc_param_grid,
        scoring='roc_auc',
        cv=3)
```

```python
display(cbc_grid_search)
```

–ü—Ä–æ—Ü–µ—Å—Å –æ–±—É—á–µ–Ω–∏—è

```python
%%time
grids = [lr_grid_search, dt_grid_search, rf_grid_search, knn_grid_search, svm_grid_search, cbc_grid_search]
for pipe in grids:
    pipe.fit(X_train,y_train)
```

```python
grid_dict = {0: 'Logistic Regression', 1: 'Decision Trees', 
             2: 'Random Forest', 3: 'K-Nearest Neighbors', 
             4: 'Support Vector Machines', 5: 'CatBoostClassifier'}
for i, model in enumerate(grids):
    print('{} Best roc_auc: {}'.format(grid_dict[i],
    model.best_score_ ))                                                                    
    print('{} Best Params: {}'.format(grid_dict[i],          
    model.best_params_))
```
